```python
class Travel_Ethiopia:
    '''we will inititalize the attributes'''
    def __init__(self, graph, initial_state, goal_state):
        self.graph = graph
        self.initial_state = initial_state.lower()
        self.goal_state = goal_state.lower()
    ''' The breadth first search algorithm '''
    def breadth_first(self):
        search = deque([self.initial_state])
        searched = set()
        count = 0
        while search:
            check_zero = search.popleft()
            if check_zero not in searched:
                if check_zero != self.goal_state:
                    count += 1
                    try:
                        search.extend(self.graph[check_zero])
                    except Exception as e:
                        print(f"you committed {e}")
                    searched.add(check_zero)
                else:
                    print(f"{count} traversed before getting in to {check_zero}")
                    print("Got It!!")
                    return check_zero.title()
                    
    ''' Now lets implement the depth first algorithm '''
    def depth_first(self, node, depth=0, visited=None):
        if visited is None:
            visited = set()

        if node == self.goal_state:
            print(f"The agent went {depth} steps down before it reached the goal")
            return node

        visited.add(node)

        for neighbor in self.graph.get(node, []):
            if neighbor not in visited:
                self.depth_first(neighbor, depth + 1, visited)
## There is a bug to be fixed
```
